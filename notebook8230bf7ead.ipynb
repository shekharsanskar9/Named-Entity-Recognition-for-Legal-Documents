```python
import numpy as np
import scipy
from scipy import special

# Print versions (correct place to get SciPy version)
print("NumPy version:", np.__version__)
print("SciPy version:", scipy.__version__)

# small numerical sanity checks
x = np.linspace(-1, 1, 5)   # points in [-1,1]
print("\nx =", x)

# classical Legendre polynomial P3(x)
P3 = special.legendre(3)    # returns a callable polynomial object
print("\nLegendre P3(x):", P3(x))

# associated Legendre P_n^m(x) — replacement for old sph_legendre use
# note: lpmv(m, n, x) expects m,n as integers (m>=0 usually)
print("\nAssociated Legendre (m=1, n=3) lpmv(1,3,x):", special.lpmv(1, 3, x))

# show that the special module exists and list some useful functions
print("\nSome available functions in scipy.special (sample):", 
      [name for name in dir(special) if name.startswith("l")][:12])
```

```python
# Cell 3: Load Dataset and Define Labels

print("Loading dataset...")
dataset = load_dataset("opennyaiorg/InLegalNER")

# Get the list of NER tags
label_list = dataset['train'].features['ner_tags'].feature.names
num_labels = len(label_list)

# Create id <-> label mappings
id2label = {i: label for i, label in enumerate(label_list)}
label2id = {label: i for i, label in enumerate(label_list)}

print("Dataset loaded.")
print(f"Labels: {label_list}")
```

```python
# Cell 4: Tokenization and Label Alignment

# We use "roberta-base". It's a strong baseline.
model_checkpoint = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint, add_prefix_space=True)

def tokenize_and_align_labels(examples):
    """
    Tokenizes text and aligns NER labels to the new subword tokens.
    Sets label to -100 for special tokens and non-first subword tokens.
    """
    tokenized_inputs = tokenizer(
        examples["tokens"], 
        truncation=True, 
        is_split_into_words=True
    )

    labels = []
    for i, label in enumerate(examples["ner_tags"]):
        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []
        for word_idx in word_ids:
            if word_idx is None:
                # Special token (e.g., [CLS], [SEP])
                label_ids.append(-100)
            elif word_idx != previous_word_idx:
                # First token of a new word
                label_ids.append(label[word_idx])
            else:
                # Subsequent subword token
                label_ids.append(-100)
            previous_word_idx = word_idx
        labels.append(label_ids)

    tokenized_inputs["labels"] = labels
    return tokenized_inputs

print("Applying tokenization and alignment to all splits...")
tokenized_datasets = dataset.map(tokenize_and_align_labels, batched=True)
print("Tokenization complete.")
```

```python
# Cell 5: Evaluation Metric Setup (Seqeval)

# Load the seqeval metric
metric = evaluate.load("seqeval")

def compute_metrics(p):
    """
    Computes Precision, Recall, F1, and Accuracy for NER.
    Also extracts and flattens per-class F1 scores.
    """
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    # Convert IDs to label strings
    true_predictions = [
        [label_list[p] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label_list[l] for (p, l) in zip(prediction, label) if l != -100]
        for prediction, label in zip(predictions, labels)
    ]

    # Compute metrics using seqeval
    results = metric.compute(predictions=true_predictions, references=true_labels)

    # Flatten results for easier logging
    flat_results = {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
        "accuracy": results["overall_accuracy"],
    }
    
    # Add per-class F1 scores
    for key, value in results.items():
        if isinstance(value, dict) and 'f1' in value:
            # e.g., "f1_PETITIONER", "f1_COURT"
            flat_results[f"f1_{key}"] = value['f1']
            
    return flat_results

print("compute_metrics function defined.")
```

```python
# Cell 6: Model & Training Configuration

# Load the model
model = AutoModelForTokenClassification.from_pretrained(
    model_checkpoint,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

# Data collator for padding
data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)

# Define output directory for persistence
output_dir = "/kaggle/working/roberta-legal-ner"

# Define Training Arguments
training_args = TrainingArguments(
    output_dir=output_dir,
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,  # 3 epochs is a good balance
    weight_decay=0.01,
    evaluation_strategy="epoch",  # Evaluate at the end of each epoch
    save_strategy="epoch",        # Save a checkpoint at the end of each epoch
    load_best_model_at_end=True,  # Load the best model (by F1) at the end
    metric_for_best_model="f1",
    report_to="none",             # Disable external reporting
)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"], # Use 'validation' for eval
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

print("Trainer initialized. Model and TrainingArguments are ready.")
```

```python
# Cell 7: Train the Model
# This will take 20-40 minutes on a T4 GPU.

print("Starting model training...")
train_result = trainer.train()
print("Training complete.")

# Save training stats
with open(f"{output_dir}/train_results.json", 'w') as f:
    json.dump(train_result.metrics, f)
```

```python
# Cell 8: Evaluate on Test Set & Save Final Model

print("Evaluating on the test set...")
test_results = trainer.evaluate(tokenized_datasets["test"])

print("\n--- Test Set Results ---")
print(test_results)
print("------------------------")

# --- PERSISTENCE ---
# Save the final model, tokenizer, and test results
final_model_path = "/kaggle/working/roberta-legal-ner-final"
trainer.save_model(final_model_path)
tokenizer.save_pretrained(final_model_path)

# Save the test results to a file
results_file_path = f"{final_model_path}/test_results.json"
with open(results_file_path, 'w') as f:
    json.dump(test_results, f, indent=4)

print(f"Final model saved to: {final_model_path}")
print(f"Test results saved to: {results_file_path}")
```

```python
# Cell 9: Load Your Trained Model

final_model_path = "/kaggle/working/roberta-legal-ner-final"
custom_ner_pipeline = pipeline(
    "ner", 
    model=final_model_path, 
    tokenizer=final_model_path, 
    aggregation_strategy="simple" # Groups subwords (e.g., Ram, ##esh -> Ramesh)
)

# nlp_blank was already created in our setup cell
print("Your custom-trained RoBERTa model is loaded.")
```

```python
# Cell 10: Qualitative Visualization (displaCy)
# This cell now only shows your model's output

# A good test sentence
text = "The Hon'ble Supreme Court of India decided the case of Ram Singh vs State of Bihar under Section 302 of IPC."

# Define colors for the entities (from the repo)
colors = {
    'COURT': "#bbabf2", 'PETITIONER': "#f570ea", "RESPONDENT": "#cdee81",
    'JUDGE': "#fdd8a5", "LAWYER": "#f9d380", 'WITNESS': "violet",
    "STATUTE": "#faea99", "PROVISION": "yellow", 'CASE_NUMBER': "#fbb1cf",
    "PRECEDENT": "#fad6d6", 'DATE': "#b1ecf7", 'OTHER_PERSON': "#b0f6a2",
    'ORG':'#a57db5','GPE':'#7fdbd4'
}
options = {"colors": colors}

# --- 1. Our Custom RoBERTa Model ---
print("--- Output from Your Custom RoBERTa Model ---")
doc_custom = nlp_blank(text) # Create a blank doc
roberta_entities = custom_ner_pipeline(text)

# We must manually add the pipeline's results to the blank SpaCy doc
ents = []
for ent in roberta_entities:
    # Create a Span for each entity
    span = doc_custom.char_span(
        ent['start'], 
        ent['end'], 
        label=ent['entity_group']
    )
    # Handle potential tokenization misalignments
    if span is not None:
        ents.append(span)
doc_custom.ents = ents

# Render the custom model's output
displacy.render(doc_custom, style="ent", jupyter=True, options=options)
```

```python
# Cell 11: Quantitative Visualization (Your Model's F1 Scores)
# This plot shows the per-class F1 score for your model

print("Plotting your model's F1 scores by entity type...")

# 1. Load your RoBERTa model's test results
results_file_path = "/kaggle/working/roberta-legal-ner-final/test_results.json"
with open(results_file_path, 'r') as f:
    roberta_results = json.load(f)

# 2. Extract per-class F1 scores
per_class_data = {
    key.replace('eval_f1_', ''): val 
    for key, val in roberta_results.items() 
    if key.startswith('eval_f1_')
}

df_per_class = pd.DataFrame(
    list(per_class_data.items()), 
    columns=['Entity', 'F1-Score']
).sort_values('F1-Score', ascending=False)

# 3. Plot the results
plt.figure(figsize=(20, 8))
sns.barplot(
    data=df_per_class,
    x="Entity",
    y="F1-Score",
    palette="viridis"
)
plt.title("Custom RoBERTa Model: Per-Class F1 Scores (Test Set)", fontsize=20)
plt.ylabel("F1-Score", fontsize=14)
plt.xlabel("Entity Type", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.show()

# 4. Print overall scores
print("\n--- Overall Model Performance ---")
print(f"Overall F1:        {roberta_results['eval_f1']:.4f}")
print(f"Overall Precision: {roberta_results['eval_precision']:.4f}")
print(f"Overall Recall:    {roberta_results['eval_recall']:.4f}")
```
